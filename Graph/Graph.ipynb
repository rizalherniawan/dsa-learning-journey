```python
g = nx.Graph()
 
g.add_edge("A","B")
g.add_edge("B","C")
g.add_edge("A","C")
g.add_edge("B","D")
g.add_edge("D","E")
g.add_edge("E","F")
g.add_edge("D","F")
g.add_edge("C","G")
g.add_edge("G","H")
g.add_edge("H","I")
g.add_edge("G","I")
g.add_edge("F","H")

nx.draw(g, with_labels = True)
```
